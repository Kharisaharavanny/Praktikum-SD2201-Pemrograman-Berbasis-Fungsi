```python
import pandas as pd

file_path = '/content/911_Calls_for_Service_(Last_30_Days).csv'
df = pd.read_csv(file_path)

filtered_df = df[
    (df['zip_code'] != '') &
    (df['neighborhood'] != '') &
    (df['totalresponsetime'] != '') &
    (df['traveltime'] != '') &
    (df['totaltime'] != '')
]

samples_neighborhood = {neighborhood: filtered_df[filtered_df['neighborhood']== neighborhood]
                        for neighborhood in filtered_df['neighborhood'].unique()}

for neighborhood, data_neighborhood in samples_neighborhood.items():
    # Menghitung rata-rata total waktu respons
    rata2_totalresponsetime = data_neighborhood['totalresponsetime'].astype(float).mean()
    # Menghitung rata-rata waktu pengiriman
    rata2_traveltime = data_neighborhood['traveltime'].astype(float).mean()
    # Menghitung rata-rata total waktu
    rata2_totaltime = data_neighborhood['totaltime'].astype(float).mean()
    print(f"Neighborhood: {neighborhood}")
    print(f"Total waktu respons rata-rata: {rata2_totalresponsetime}")
    print(f"Waktu pengiriman rata-rata   : {rata2_traveltime}")
    print(f"Total waktu rata-rata        : {rata2_totaltime}")
    print()
```

```python
import pandas as pd

file_path = '/content/911_Calls_for_Service_(Last_30_Days).csv'
df = pd.read_csv(file_path)

filtered_df = df[
    (df['zip_code'] != '') &
    (df['neighborhood'] != '') &
    (df['totalresponsetime'] != '') &
    (df['traveltime'] != '') &
    (df['totaltime'] != '')
]

rata2_totalresponsetime = filtered_df['totalresponsetime'].astype(float).mean()
rata2_traveltime = filtered_df['traveltime'].astype(float).mean()
rata2_totaltime = filtered_df['totaltime'].astype(float).mean()

print("Rata-rata waktu:")

print(f"Total waktu respons rata-rata: {rata2_totalresponsetime}")
print(f"Waktu pengiriman rata-rata   : {rata2_traveltime}")
print(f"Total waktu rata-rata        : {rata2_totaltime}")
```
